```python
import pandas as pd
from spicy import stats
```

```python
dataset_path = './datasets/consolidated/consolidated.csv'

data = pd.read_csv(
    filepath_or_buffer=dataset_path,
)

data['date'] = pd.to_datetime(data['date']).dt.date

data.drop(columns=['date'], inplace=True)

data
```

```python
# Iterar sobre las columnas y realizar la prueba de Anderson-Darling
for columna in data.columns:
    resultado = stats.anderson(data[columna])

    # Obtener el valor crítico y estadístico
    estadistica = resultado.statistic
    valor_critico = resultado.critical_values[2]  # Nivel de significancia del 5% (índice 2)

    # Imprimir los resultados para cada columna
    print(f"Columna: {columna}")
    print(f"Nivel de significancia: 0.05")
    print(f"Valor crítico: {valor_critico}")
    print(f"Estadística de prueba: {round(estadistica, 4)}")

    # Comparación con el valor crítico correspondiente al nivel de significancia
    if estadistica > valor_critico:
        print("Se rechaza la hipótesis nula: los datos no siguen una distribución normal")
    else:
        print("No se rechaza la hipótesis nula: los datos podrían seguir una distribución normal")
    print()
```
